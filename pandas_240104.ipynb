```python
table_data3 = {'봄':  [256.5, 264.3, 215.9, 223.2, 312.8],
              '여름': [770.6, 567.5, 599.8, 387.1, 446.2],
              '가을': [363.5, 231.2, 293.1, 247.7, 381.6],
              '겨울': [139.3, 59.9, 76.9, 109.1, 108.1]}
columns_list = ['봄', '여름', '가을', '겨울']
index_list = ['2012', '2013', '2014', '2015', '2016']

df3 = pd.DataFrame(table_data3, columns = columns_list, index = index_list)
df3
```

```python
print(df3.mean())
print(df3.max())
print(df3.min())
print(df3.std())
```

```python
iris = sns.load_dataset('iris')

#sepal_length의 길이가 7cm 이상인 것만 조회
iris.loc[iris['sepal_length'] >=7,:]
```

```python
#species가 virginica인것만 조회
iris.loc[iris['species'] =='virginica',:].head(3)
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# embark_town 컬럼에서 Southampton 거주민 데이터
titanic.loc[titanic['embark_town'] =='Southampton']
```

```python
#age 평균 구하고, 평균 이상인 데이터만 조회
titanic.loc[titanic['age'] >= titanic['age'].mean()]
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#tota_bill의 중간값 이하인 것만 조회
print(tips.loc[tips['total_bill'] <=tips['total_bill'].median()].head(2))

#day가 sun인데이터만 조회
print(tips.loc[tips['day']=='Sun'].head(2))

# tip의 평균보다 크면서, time 이 Dinner인 데이터만조회
print(tips.loc[(tips['tip'] > tips['tip'].mean()) & (tips['day']=='Sun')].head())
```
