```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    'Class1' : [95,92,98,100],
    'Class2' : [91,93,97,99]})

df2 = pd.DataFrame({
    'Class1' : [87,89],
    'Class2' : [85,90]})

result = pd.concat([df1,df2],ignore_index=True)
#https://pandas.pydata.org/pandas-docs/version/1.4/reference/api/pandas.concat.html#pandas.concat
result
```

```python
df3 = pd.DataFrame({
    'Class1' : [96,83]})

pd.concat([result,df3], ignore_index = True)
```

```python
df_A_B = pd.DataFrame({'판매월': ['1월', '2월', '3월', '4월'],
                       '제품A': [100, 150, 200, 130],
                       '제품B': [90, 110, 140, 170]})

df_C_D = pd.DataFrame({'판매월': ['1월', '2월', '3월', '4월'],
                       '제품C': [112, 141, 203, 134],
                       '제품D': [90, 110, 140, 170]})
```

```python
df_left = pd.DataFrame({'key':['A','B','C'], 'left': [1, 2, 3]})
df_right = pd.DataFrame({'key':['A','B','D'], 'right': [4, 5, 6]})
display(df_left)
display(df_right)
```

```python
df_left.merge(df_right, how = 'left', on='key')
```

```python
df_left.merge(df_right, how = 'right', on='key')
```

```python
df_left.merge(df_right, how = 'outer', on='key') #=FULL JOIN
```

```python
df_left.merge(df_right, how = 'inner', on='key') #교집합
```

```python
#How to combine data from multiple tables
air_quality_no2 = pd.read_csv("data/air_quality_no2_long.csv",
                              parse_dates=True)

air_quality_no2 = air_quality_no2[["date.utc", "location",
                                   "parameter", "value"]]
air_quality_no2.head()
```

```python
air_quality_pm25 = pd.read_csv("data/air_quality_pm25_long.csv",
                               parse_dates=True)


air_quality_pm25 = air_quality_pm25[["date.utc", "location",
                                     "parameter", "value"]]


air_quality_pm25.head()
```

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)

air_quality.head()
```

```python
air_quality = air_quality.sort_values("date.utc", ignore_index=True)
air_quality.head()
```

```python
air_quality_ = pd.concat([air_quality_pm25, air_quality_no2], keys=["PM25", "NO2"])

air_quality_.head()
```

```python
stations_coord = pd.read_csv("data/air_quality_stations.csv")

stations_coord.head()
```

```python
air_quality = pd.merge(air_quality, stations_coord, how="left", on="location")

air_quality.head()
```

```python
air_quality_parameters = pd.read_csv("data/air_quality_parameters.csv")

air_quality_parameters.head()
```

```python
air_quality = pd.merge(air_quality, air_quality_parameters,
                       how='left', left_on='parameter', right_on='id')


air_quality.head()
```
